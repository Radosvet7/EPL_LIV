## Exploratory Data Analysis

```python
#importing libs
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
warnings.filterwarnings("ignore")
from IPython.display import Audio, display
sound = 'EPL.mp3'
Audio(url=sound, autoplay=True)
```

```python
#loading data
url = 'https://raw.githubusercontent.com/Radosvet7/EPL_LIV/main/results.csv'
df = pd.read_csv(url, encoding='windows-1252')
df.head(5)
```

```python
#removing blank data from dataset
df = df[df['AR'].notnull()]
df.head(5)
```

```python
LIV_df = pd.concat([df[df['HomeTeam']=='Liverpool'], df[df['AwayTeam']=="Liverpool"]]).sort_index()
cols = ["Season","HomeTeam","AwayTeam","FTR","FTHG","FTAG",'DateTime', 'HS', 'AS', 'HST', 'AST', 'HTR']
LIV_df = LIV_df[cols]
LIV_df['Date'] = pd.to_datetime(LIV_df['DateTime'])
LIV_df = LIV_df.drop(columns=['DateTime'])
LIV_df.head()

def win_loss(x):
    if(x[1]=="Liverpool"):
        if(x[3]=="A"):
            return "L"
        elif(x[3]=="H"):
            return "W"
    elif(x[2]=="Liverpool"):
        if(x[3]=="A"):
            return "W"
        elif(x[3]=="H"):
            return "L"
    return "D"

results = LIV_df.apply(win_loss, axis = 1 )
LIV_df['Results'] = results
LIV_df.to_csv("output.csv", index=False)
```

```python
#creating variables for Goals scored(GS), Goals against(GA), Points(Pts), Goal Difference(GD)
dict1 = {'W' : 3, 'D' : 1, 'L' : 0}
LIV_df['Pts'] = LIV_df['Results'].map(dict1)

cond = [(LIV_df['HomeTeam'] == 'Liverpool'),
        (LIV_df['AwayTeam'] == 'Liverpool')]

GS = [LIV_df['FTHG'], LIV_df['FTAG']]
GA = [LIV_df['FTAG'], LIV_df['FTHG']]

LIV_df['GS'] = np.select(cond, GS)
LIV_df['GA'] = np.select(cond, GA)
LIV_df['GD'] = LIV_df['GS'] - LIV_df['GA']
LIV_df.head()
```

```python
#managers reign
conditions = [(LIV_df['Date']<= '2004-05-24'),
              (LIV_df['Date']>='2004-06-16') & (LIV_df['Date']<='2010-06-03'),
              (LIV_df['Date']>='2010-07-01') & (LIV_df['Date']<='2011-01-08'),
              (LIV_df['Date']>='2011-01-09') & (LIV_df['Date']<='2012-05-16'),
              (LIV_df['Date']>='2012-06-01') & (LIV_df['Date']<='2015-10-04'),
              (LIV_df['Date']>='2015-10-05')]
values = ['GÃ©rard Houllier', 'Rafael BenÃ­tez','Roy Hodgson', 'Kenny Dalglish', 'Brendan Rodgers', 'JÃ¼rgen Klopp']
LIV_df['Manager'] = np.select(conditions, values)
LIV_df.head(5)
```

```python
#manager performance overview
ord = ['W', 'D', 'L']
result = pd.crosstab(index= LIV_df['Manager'],
                    columns = LIV_df['Results'],
                    values = LIV_df['Results'],
                    aggfunc = 'count',
                    normalize = 'index') * 100
result[ord].sort_values(by=['W'], ascending=False).round(2).astype(str) + '%'
```

```python
#average point per game won by each manager
# Assuming you already have a DataFrame named LIV_df

# Group by 'Season' and 'Manager' and calculate the average points per game
avg = LIV_df.groupby(['Season', 'Manager']).apply(lambda x: x['Pts'].sum() / len(x))

# Convert the result to a DataFrame
avg_df = avg.reset_index(name='Avg_Pts_Per_Game')

# Sort the DataFrame by 'Avg_Pts_Per_Game' in descending order
avg_df_sorted = avg_df.sort_values(by='Avg_Pts_Per_Game', ascending=False)

# Display the sorted DataFrame
print(avg_df_sorted)
```

```python
all_sns = LIV_df['Season'].unique()
data = {'Season': [], 'Team': [], 'Pts': []}
for sns in all_sns:
    df1 = df[df['Season']==sns]
    all_teams = df1['HomeTeam'].unique()
    for team in all_teams:
        df2 = pd.concat([df1[df1['HomeTeam']== team], df1[df1['AwayTeam']== team]]).sort_index()
        df2.loc[(df2['HomeTeam'] == team)&(df2['FTR']=='H'), 'Pts'] = 3
        df2.loc[(df2['HomeTeam'] == team)&(df2['FTR']=='D'), 'Pts'] = 1
        df2.loc[(df2['AwayTeam'] == team)&(df2['FTR']=='A'), 'Pts'] = 3
        df2.loc[(df2['AwayTeam'] == team)&(df2['FTR']=='D'), 'Pts'] = 1
        sum = df2['Pts'].sum()
        data['Season'].append(sns)
        data['Team'].append(team)
        data['Pts'].append(sum)
df = pd.DataFrame.from_dict(data)
df['Position'] = df.groupby('Season')['Pts'].rank(ascending=False, method='min').astype(int)
df.to_csv("test.csv", index=False)
```

```python
linechart = df[df['Team'].isin(['Man United', 'Liverpool'])]
data = linechart[['Season', 'Team', 'Pts']]
data['Season'] = data['Season'].apply(lambda x: x[2:4] + '-' + x[-2:])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
custom_palette = {'Man United': 'black', 'Liverpool': 'red'}
custom_style = {'Man United': '--', 'Liverpool': '-.'}
sns.set_style('ticks')
sns.lineplot(x='Season', y='Pts', hue='Team', data=data, marker='*', sort=False, palette = custom_palette)
# Add labels and title
plt.xlabel('Season')
plt.ylabel('Points')
plt.title('Points Over Seasons')
plt.xticks(rotation=45)

manager_arrivals = {'JÃ¼rgen Klopp': '15-16'}

for manager, arrival_season in manager_arrivals.items():
    season_data = data[data['Season'] == arrival_season]
    
    if not season_data.empty:
        pts = season_data['Pts'].values[1]
        # Apply split, get the first part, fill NaN values with 0, and convert to int
        season_numeric = season_data['Season'].str.split('-').str[0].fillna(0).astype(int)
        
        plt.annotate(f'{manager} arrival', 
                     xy=(season_numeric.values[0], pts), 
                     xytext=(season_numeric.values[1], pts - 5), 
                     ha='center',
                     arrowprops=dict(facecolor='red', edgecolor='green', linewidth = 1, shrink=0.02))

# Show the plot
plt.grid(True)
plt.legend()
plt.show()
```
